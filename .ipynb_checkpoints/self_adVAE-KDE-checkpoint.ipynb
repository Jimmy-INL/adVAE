```python
from __future__ import print_function

import torch
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.metrics import (precision_recall_curve, auc,average_precision_score,
                             roc_curve,precision_score, recall_score,  f1_score)
import statsmodels.api as sm
from utils.plot_culve import plot_ROC,plot_PRC
import time
from load_data.tabular import load_tab_data
import models 
import seaborn as sns
#import torch.utils.data
```

```python
SEED=1
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
os.environ['PYTHONHASHSEED'] = str(SEED)
torch.cuda.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic=True
torch.backends.cudnn.benchmark = False
```

```python
def reparameterize_Ltimes(mu, logvar,L):
    std = torch.exp(0.5*logvar)
    std = std.unsqueeze(1).repeat(1,L, 1)
    mu = mu.unsqueeze(1).repeat(1,L, 1)
    eps = torch.randn_like(std)
    return eps.mul(std).add_(mu)
```

```python
def trainset_anomaly_score(x_train,encoder,decoder):
    print('computing anomaly score in training dataset...')
    encoder.eval()
    decoder.eval()
    L=1000
    #x_recon_error=torch.FloatTensor(x_train.size()[0]).zero_()
    with torch.no_grad():
        mu_train, logvar_train=encoder(x_train)
        z_train=reparameterize_Ltimes(mu_train, logvar_train,L)
        x_recon=decoder(z_train)
        res=x_recon-x_train.unsqueeze(1).repeat(1,L,1)
        res=torch.mean(res, 1, keepdim=False)
        spe=torch.diag(torch.mm(res,torch.transpose(res, 0, 1)))
    return spe.cpu().numpy()
```

```python
def testset_anomaly_score(x_test,encoder,decoder):
    print('computing anomaly score in testing dataset...')
    encoder.eval()
    decoder.eval()
    L=1000
    with torch.no_grad():
        mu_test, logvar_test=encoder(x_test)
        z_test=reparameterize_Ltimes(mu_test, logvar_test,L)       
        x_recon=decoder(z_test)
        res=x_recon-x_test.unsqueeze(1).repeat(1,L,1)
        res=torch.mean(res, 1, keepdim=False)
        spe=torch.diag(torch.mm(res,torch.transpose(res, 0, 1)))
    return spe.cpu().numpy()
```

```python
def learning_threshold(train_recon_error,alpha):
    dens = sm.nonparametric.KDEUnivariate(train_recon_error.astype(np.float))
    bound = np.linspace(min(train_recon_error),max(train_recon_error),1024)
    dens.fit(bw='silverman',gridsize=1024)
    threshold=bound[min(np.where(dens.cdf>alpha)[0])]
    return threshold
```

```python
def anomaly_detect(recon_error,threshold):
    ano_pred=np.zeros(recon_error.shape[0])
    for i in range(recon_error.shape[0]):
        if recon_error[i]>threshold:
            ano_pred[i]=1
    return ano_pred
```

```python
def show_pre_recall_f1(test_recon_error,test_label,threshold):
    y_pred=test_recon_error.copy()
    larger_idx=np.where(y_pred>=threshold)
    lower_idx=np.where(y_pred<threshold)
    y_pred[lower_idx[0]]=0
    y_pred[larger_idx[0]]=1
    print('precision:',precision_score(test_label,y_pred))
    print('recall:',recall_score(test_label,y_pred))
    print('f1 score:',f1_score(test_label,y_pred))
```

```python
def ploting_curves_with_thresholds(dataset_name,test_labels,train_recon_error,test_recon_error,KDEthreshold,Percent_threshold):
    threshold=np.linspace(min(train_recon_error),max(train_recon_error),100)
    f1score=np.zeros(threshold.shape[0])
    Rscore=np.zeros(threshold.shape[0])
    Pscore=np.zeros(threshold.shape[0])
    #KDE95threshold=learning_threshold(train_recon_error,0.95)
    for i in range(threshold.shape[0]):
        Rscore[i]=recall_score(test_labels, anomaly_detect(test_recon_error,threshold=threshold[i]))
        f1score[i]=f1_score(test_labels, anomaly_detect(test_recon_error,threshold=threshold[i]))
        Pscore[i]=precision_score(test_labels, anomaly_detect(test_recon_error,threshold=threshold[i]))
        
    idx = (np.abs(threshold-KDEthreshold)).argmin() 
    Percent_idx = (np.abs(threshold-Percent_threshold)).argmin()
    #plt.title(dataset_name)
    plt.ylim(0, 1.05)
    plt.plot(threshold,f1score,color='g',label='F1 Score')
    plt.plot(threshold,Rscore,color='m',label='Recall')
    plt.plot(threshold,Pscore,color='b',label='Precision')
    plt.vlines(KDEthreshold, 0, Rscore[idx],colors='r',linestyles='dashed',label='KDE threshold')
    plt.vlines(Percent_threshold, 0, Rscore[Percent_idx],colors='k',linestyles='dashdot',label='Uniform threshold')
    plt.plot(KDEthreshold,f1score[idx],'go',label="")
    plt.plot(KDEthreshold,Rscore[idx],'mo',label="")
    plt.plot(KDEthreshold,Pscore[idx],'bo',label="")
    print('KDE-F1',f1score[idx])
    print('KDE-R',Rscore[idx])
    print('KDE-P',Pscore[idx])
    plt.plot(Percent_threshold,f1score[Percent_idx],'go',label="")
    plt.plot(Percent_threshold,Rscore[Percent_idx],'mo',label="")
    plt.plot(Percent_threshold,Pscore[Percent_idx],'bo',label="")
    print('P-KDE-F1',f1score[Percent_idx])
    print('P-KDE-R',Rscore[Percent_idx])
    print('P-KDE-P',Pscore[Percent_idx])
    plt.legend(loc=4)
#     plt.plot(KDE95threshold,f1score[idx95],'go',label="")
#     plt.plot(KDE95threshold,Rscore[idx95],'mo',label="")
#     plt.plot(KDE95threshold,Pscore[idx95],'bo',label="")
    plt.ylabel('Metric values')
    plt.xlabel('Thresholds')
    plt.savefig(dataset_name+'_KDE.jpg',dpi=600)
    plt.show()
```

```python
def show_results(dataset_name):
    data_path='./datasets/'
    save_path='D:/GitHub/selfADVAE-AD/checkpoints'
    device='cuda'
    batch_size=32
    n_jobs_dataloader=0
    
    #load_dataset
    train_data,train_label,test_data,test_label,sample_dim,rep_dim=load_tab_data(data_path=data_path, dataset_name=dataset_name)
    scaler=MinMaxScaler().fit(train_data)
    train_data_scaled=scaler.transform(train_data) 
    test_data_scaled=scaler.transform(test_data) 
    x_train=torch.cuda.FloatTensor(train_data_scaled)
    x_test=torch.cuda.FloatTensor(test_data_scaled)
    
    # build model
    encoder = models.Encoder(sample_dim=sample_dim,rep_dim=rep_dim).to(device)
    decoder = models.Decoder(sample_dim=sample_dim,rep_dim=rep_dim).to(device)
    
    #load a well-trained adVAE model
    net_name=dataset_name+'selfADVAE' 
    encoder.load(save_path+'/'+net_name+'encoder')
    decoder.load(save_path+'/'+net_name+'decoder')
    
    #computing anomaly score
    train_recon_error=trainset_anomaly_score(x_train,encoder,decoder)
    start_time = time.time()
    test_recon_error=testset_anomaly_score(x_test,encoder,decoder)
    train_time = time.time() - start_time
    print('testing time: %.3f' % train_time)
    print('min anomaly score: %.3f' % min(train_recon_error))
    print('max anomaly score: %.3f' % max(train_recon_error))
    
    #train_recon_error_=(train_recon_error-min(train_recon_error))/(max(train_recon_error)-min(train_recon_error))
    Percent_threshold=np.sort(train_recon_error)[int(0.9*train_recon_error.shape[0])+1]
    #learning_threshold
    KDE_threshold=learning_threshold(train_recon_error,0.9)
    print('KDEthreshold',KDE_threshold)
    print('Percentthreshold',Percent_threshold)
    ploting_curves_with_thresholds(dataset_name,test_label,train_recon_error,test_recon_error,KDE_threshold,Percent_threshold)
#     show_pre_recall_f1(test_recon_error,test_label,threshold)
    torch.cuda.empty_cache()
#     # plot PRC and ROC
#     plot_PRC(test_label, test_recon_error)
#     plot_ROC(test_label, test_recon_error)
    return train_recon_error
```
